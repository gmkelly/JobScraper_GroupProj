```python
import pandas as pd

#Load the national data csv into a dataframe.
df_nationaldata = pd.read_csv("job_data_national.csv", thousands=',') 
#Preview the national dataframe.
#df_nationaldata.head()

#Load the state data csv into a dataframe.
df_statedata = pd.read_csv("job_data_states.csv", thousands=',')
#Preview the state dataframe.
#df_statedata.head()

#Combine the two dataframes.
df_nationalandstate = pd.concat([df_nationaldata, df_statedata], ignore_index=True)
```

```python
#Look at the combined dataframe to make sure it worked.
df_nationalandstate
```

```python
#Use regex to convert the currency columns to float datatype.
df_nationalandstate[df_nationalandstate.columns[7:10]] = df_nationalandstate[df_nationalandstate.columns[7:10]].replace('[\$,]', '', regex=True).astype(float)
```

```python
#Convert percentage columns to floats.
df_nationalandstate['EmploymentRSE'] = df_nationalandstate['EmploymentRSE'].str.rstrip('%').astype('float')
df_nationalandstate['MeanWageRSE'] = df_nationalandstate['MeanWageRSE'].str.rstrip('%').astype('float')
```

```python
#Check the datatypes of the dataframe to ensure they are appropriate.
df_nationalandstate.dtypes
```

```python
#Drop rows with all nulls (i.e., empty rows.)
df_nationalandstate.dropna(axis=0, how='all')
```

```python
#Analysis Queries
#1.1 Show records for one job (‘Computer and Information Systems Managers’) across all states and national records.
df_nationalandstate[(df_nationalandstate.OccupationTitle == 'Computer and Information Systems Managers')]
```

```python
#1.2 Show all records for ‘Computer and Mathematical Occupations’ jobs (i.e., whose occupation code begins with 15) 
df_nationalandstate[df_nationalandstate.OccupationCode.str.startswith(('15'))]
```

```python
# 1.3 Show all records where ‘Mean hourly wage’ is greater than or equal to $135. Order by highest mean hourly wage. 
df_nationalandstate[(df_nationalandstate.MeanHrWage >= 135)].sort_values(by=['MeanHrWage'], ascending=False)
```

```python
#1.4 Show all records where ‘Annual mean wage’ is less than $20,000. Order by lowest annual mean wage. 
df_nationalandstate[(df_nationalandstate.AnnualMeanWage < 20000)].sort_values(by=['AnnualMeanWage'])
```

```python
#1.5 Show all records where ‘Location quotient’ is less than 0.03 (indicating a job that is extremely rare in the state compared to the national average). Order by rarest first. 
df_nationalandstate[(df_nationalandstate.LocationQuotient < 0.03)].sort_values(by=['LocationQuotient'])
```

```python
#1.6 Show all records where ‘Location quotient’ is higher than 50 (indicating a job that is extremely common in the state compared to the national average). Order by most least rare first. 
df_nationalandstate[(df_nationalandstate.LocationQuotient > 30)].sort_values(by=['LocationQuotient'], ascending=False)
```

```python
# 2.1 Find records (i.e., states) where the job is ‘Computer and Information Systems Managers’ and the location quotient is higher than 1.5 (indicating states where Information Systems Managers are more common than the national average). Sort by most common first. 
df_nationalandstate[(df_nationalandstate.OccupationTitle=='Computer and Information Systems Managers') & (df_nationalandstate.LocationQuotient > 1.5)].sort_values(by=['LocationQuotient'], ascending=False)
```

```python
# 2.2 Find states where the job is ‘Computer and Information Systems Managers’ and the location quotient is less than 0.45 (indicating states where Computer and Information Systems Managers are less common than the national average). Sort by rarest first. 
df_nationalandstate[(df_nationalandstate.OccupationTitle=='Computer and Information Systems Managers') & (df_nationalandstate.LocationQuotient < 0.45)].sort_values(by=['LocationQuotient'])
```

```python
# 2.3 Find jobs that are much more common in Louisiana than the national average (i.e., where state = Louisiana and where location quotient is higher than 10). Sort by most common first. 
df_nationalandstate[(df_nationalandstate.NationOrState=='Louisiana') & (df_nationalandstate.LocationQuotient > 10)].sort_values(by=['LocationQuotient'], ascending=False)
```

```python
# 2.4 Find jobs that are much less common in Louisiana than the national average (i.e., where state = Louisiana and where location quotient is lower than 0.1). Sort by rarest first. 
df_nationalandstate[(df_nationalandstate.NationOrState=='Louisiana') & (df_nationalandstate.LocationQuotient < 0.1)].sort_values(by=['LocationQuotient'])
```

```python
#2.5 Find states where Bartenders make more than $30,000 a year on average. Sort by highest wage.
df_nationalandstate[(df_nationalandstate.OccupationTitle == 'Bartenders') & (df_nationalandstate.AnnualMeanWage > 30000)].sort_values(by=['AnnualMeanWage'], ascending=False)
```

```python
#3.1 Find jobs in the ‘Computer and Mathematical Occupations’ field (i.e. Starting with code 15) that make more than $75,000 per year in Louisiana. Sort by highest wage.
df_nationalandstate[(df_nationalandstate.OccupationCode.str.startswith(('15'))) & (df_nationalandstate.NationOrState == 'Louisiana') & (df_nationalandstate.AnnualMeanWage > 75000)].sort_values(by=['AnnualMeanWage'], ascending=False)
```

```python
#3.2: Find jobs 1) at the national level only that 2) have less than 5000 people employed and 3) make more than $100,000 on average annually. Order by lowest number employed.
df_nationalandstate[(df_nationalandstate.NationOrState == 'National') & (df_nationalandstate.Employment < 5000) & (df_nationalandstate.AnnualMeanWage > 100000)].sort_values(by=['Employment'])
```

```python
#3.3: Find jobs 1) at the national level only that 2) have more than 4,000,000 people employed and 3) make less than $30,000 on average annually. Order by highest number employed.
df_nationalandstate[(df_nationalandstate.NationOrState == 'National')& (df_nationalandstate.Employment > 4000000) &(df_nationalandstate.AnnualMeanWage < 30000)].sort_values(by=['Employment'], ascending=False)
```

```python
#4.1: Create a new column that calculates the wage a worker in each job would receive each month, based on the ‘AnnualMeanWage’ (i.e., divide AnnualMeanWage by 12). Round to 2 decimal places.
df_nationalandstate['MonthlyWage'] = round(df_nationalandstate['AnnualMeanWage']/12, 2)

#View the dataframe to see the new column.
df_nationalandstate
```

```python
#5.1: Draw a horizontal bar graph of 'Computer and Mathematical Occupations' average salaries in Louisiana. Sort highest wage first.
df_nationalandstate[(df_nationalandstate.OccupationCode.str.startswith(('15'))) & (df_nationalandstate.NationOrState == 'Louisiana')].sort_values('AnnualMeanWage').plot.barh(x='OccupationTitle', y='AnnualMeanWage', legend=False)
```

```python
#7.4: Draw a scatter plot of the mean hourly wage vs. the median hourly wage for Computer and Mathematical Occupations.
df_nationalandstate[(df_nationalandstate.OccupationCode.str.startswith(('15')))].plot.scatter(x='MeanHrWage', y='MedianHrWage')
```

```python
#7.5: Compare annual mean wage with employment per 1000 people, for jobs with a below-average annual mean wage.
df_nationalandstate[df_nationalandstate.AnnualMeanWage<(0.5*(df_nationalandstate.AnnualMeanWage.mean()))].plot.scatter(x='EmploymentPer1000', y='AnnualMeanWage', color='purple', marker='*')
```
